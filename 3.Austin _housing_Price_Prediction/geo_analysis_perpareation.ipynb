```python
import pandas as pd

from warnings import filterwarnings
filterwarnings('ignore')
```

```python
# Load the data
input_file_path = './source/austinHousingData.csv'
raw_data = pd.read_csv(input_file_path, encoding='utf-8', index_col=None)
raw_data.head()
```

```python
cols_to_keep = [
    'zipcode',
    'latestPrice',
    'latitude',
    'longitude',
    'propertyTaxRate',
    'yearBuilt',
    'lotSizeSqFt',
    'livingAreaSqFt',
    'latest_saledate',
    'numOfPrimarySchools',
    'numOfMiddleSchools',
    'numOfHighSchools',
    'avgSchoolDistance',
    'avgSchoolRating',
    'avgSchoolSize',
    'MedianStudentsPerTeacher',
    'numOfBathrooms',
    'numOfBedrooms',
    'numOfStories'
]

house_df=raw_data[cols_to_keep]
house_df
```

```python
# Create a copy of the DataFrame
house_df01 = house_df.copy()

# 使用 to_dict(orient='records') 将 DataFrame 转换为按行分组的字典列表
dict_list = house_df01.to_dict(orient='records')

import json
# 将字典列表转换为 JSON 格式
json_data = json.dumps(dict_list, indent=4)

# Save the JSON data to a file
with open('./statics/data/housing_data_details.json', 'w', encoding='utf-8') as f:
    f.write(json_data)  # 修正此处的变量名
```

```python
house_df['zipcode'].unique()
```

```python
# Add new column: price per square feat
house_df['price_per_sqft'] = house_df['latestPrice'] / house_df['livingAreaSqFt']

# Add new column: house age
house_df['latest_saledate'] = pd.to_datetime(house_df['latest_saledate'])
house_df['house_age'] = house_df['latest_saledate'].dt.year - house_df['yearBuilt']

#review
house_df.head()
```

```python
house_df.columns
```

```python
# Calculate the average values for each zipcode
grouped_df = house_df.groupby('zipcode').agg({
    'latestPrice': 'mean',          # Average total price
    'price_per_sqft': 'mean',       # Average price per square foot
    'house_age': 'mean',            # Average house age
    'avgSchoolRating': 'mean',      # Average school rating
    'avgSchoolSize': 'mean'         # Average school size
}).reset_index()

# Round the calculated averages to the specified precision
grouped_df['latestPrice'] = grouped_df['latestPrice'].round(2)            # Round to 2 decimal places
grouped_df['price_per_sqft'] = grouped_df['price_per_sqft'].round(2)      # Round to 2 decimal places
grouped_df['house_age'] = grouped_df['house_age'].round(1)                # Round to 1 decimal place
grouped_df['avgSchoolRating'] = grouped_df['avgSchoolRating'].round(2)    # Round to 2 decimal places
grouped_df['avgSchoolSize'] = grouped_df['avgSchoolSize'].astype(int)     # Convert to integer (no decimals)

# Rename columns to reflect they are averages
grouped_df.rename(columns={
    'latestPrice': 'avg_latestPrice',
    'price_per_sqft': 'avg_price_per_sqft',
    'house_age': 'avg_house_age',
    'avgSchoolRating': 'avg_school_rating',
    'avgSchoolSize': 'avg_school_size'
}, inplace=True)

grouped_df.head()
```

```python
grouped_df.to_csv('./statics/housing_data_grouped.csv')
grouped_df.to_json('./statics/housing_data_grouped.json')
```
